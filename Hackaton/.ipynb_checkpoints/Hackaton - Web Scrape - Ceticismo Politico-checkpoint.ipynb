```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from tqdm import tqdm
import re
```

```python
def parse_url(url, main_tag, main_tag_id):
    r = requests.get(url)
    html = BeautifulSoup(r.text, 'html5lib')
    posts = html.find_all(main_tag, main_tag_id)
    
    return posts
```

```python
def get_posts(url, pages=1):
    post_data = []
    
    if pages < 1:
        return 'Página não pode ser menor que 1' 
    elif pages == 1:
        r = requests.get(url)
        html = BeautifulSoup(r.text, 'html5lib')
        posts = html.find_all('div', 'item-details')

        for i, post in enumerate(posts):
            title = post.find('h1', 'entry-title')
            date = post.find('div', 'td-module-meta-info')
            excerpt = post.find('div', 'td-post-text-content')
                    
            if (title is not None) & (date is not None) & (excerpt is not None):
                excerpt_p = ''
                for paragraph in excerpt.find_all('p', recursive=False):
                    excerpt_p = excerpt_p + paragraph.get_text()
                
                post_data.append([title.find('a').get_text(), title.find('a').get('href'), date.find('time', 'entry-date').get('datetime'), excerpt_p])
    else:
        p = 2
        while p <= pages:
            r = requests.get(url + 'page/' + str(p) + '/')
            html = BeautifulSoup(r.text, 'html5lib')
            posts = html.find_all('div', 'item-details')

            for i, post in tqdm(enumerate(posts)):
                title = post.find('h1', 'entry-title')
                date = post.find('div', 'td-module-meta-info')
                excerpt = post.find('div', 'td-post-text-content')

                if (title is not None) & (date is not None) & (excerpt is not None):
                    excerpt_p = ''
                    for paragraph in excerpt.find_all('p', recursive=False):
                        excerpt_p = excerpt_p + paragraph.get_text()
                    post_data.append([title.find('a').get_text(), title.find('a').get('href'), date.find('time', 'entry-date').get('datetime'), excerpt_p])
                    
            p = p + 1
    return np.array(post_data)
```

```python
posts = get_posts('https://ceticismopolitico.com/', pages=675)
```

```python
dataframe = pd.DataFrame(posts, columns=['title', 'url', 'date', 'summary'])
```
